```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Inserire il nome del dataset da analizzare
dataset_name = "synthetic"

folder_result = dataset_name+"_dataset_result/"

if not os.path.isdir(folder_result):
    os.makedirs(folder_result)

ageList=['(0-2)', '(4-6)', '(8-12)', '(15-20)', '(25-32)', '(38-43)', '(48-53)', '(60-100)']
genderList=['m','f']
ethnicitiesList = ['white', 'black', 'asian', 'indian', 'others']
emotionsList = ["angry", "disgusted", "fearful", "happy", "neutral", "sad", "surprised"]

df = pd.read_csv("features_"+dataset_name+"_dataset_results.csv")
df.head()
```

# Helper Function #

```python
def printPieChart(data, labels, title, save=False):
    plt.figure(figsize =(10, 7))
    plt.title(title)
    plt.pie(data, labels = labels, autopct='%1.1f%%')
    if save:
        plt.savefig(folder_result+title+'.jpeg', dpi=300, bbox_inches="tight")
    else:
        plt.show()

def printDistribution(distribution, title, xlabel, save=False):
    distribution.T.plot(kind="bar", )
    plt.xlabel(xlabel)
    plt.ylabel("Total")
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.xticks(rotation=45)
    plt.title(title)
    if save:
        plt.savefig(folder_result+title+'.jpeg', dpi=300, bbox_inches="tight")
    
        
```

# Ethnicities distribution #

```python
distribution=pd.Series(0, index=ethnicitiesList)
distributionByAge=pd.DataFrame(0, index=ethnicitiesList, columns=ageList)
distributionByGender=pd.DataFrame(0, index=ethnicitiesList, columns=genderList)

for indexEth, eth in enumerate(ethnicitiesList):
    group = df.loc[df["ethnicity"] == indexEth]
    distribution[eth] = group.shape[0]
    for indexAge, age in enumerate(ageList):
        distributionByAge[age][eth] = group.loc[group["age"] == indexAge].shape[0]
    for indexGen, gen in enumerate(genderList):
        distributionByGender[gen][eth] = group.loc[group["gender"] == indexGen].shape[0]

print(distributionByAge)
print(distributionByGender)
printPieChart(distribution, ethnicitiesList, "Ethnicities distribution - "+dataset_name+" dataset", save=True)
printDistribution(distributionByAge, "Ethnicities distribution by ages - "+dataset_name+" dataset", "Age", save=True)
printDistribution(distributionByGender, "Ethnicities distribution by gender - "+dataset_name+" dataset", "Gender", save=True)
```

# Emotions distribution #

```python
distribution=pd.Series(0, index=emotionsList)
distributionByEmotion=pd.DataFrame(0, index=emotionsList, columns=ageList)
distributionByGender=pd.DataFrame(0, index=emotionsList, columns=genderList)

for indexEmo, emo in enumerate(emotionsList):
    group = df.loc[df["emotion"] == indexEmo]
    distribution[emo] = group.shape[0]
    for indexAge, age in enumerate(ageList):
        distributionByEmotion[age][emo] = group.loc[group["age"] == indexAge].shape[0]
    for indexGen, gen in enumerate(genderList):
        distributionByGender[gen][emo] = group.loc[group["gender"] == indexGen].shape[0]

print(distributionByEmotion)
print(distributionByGender)
printPieChart(distribution, emotionsList, "Emotions distribution - "+dataset_name+" dataset", save=True)
printDistribution(distributionByEmotion, "Emotions distribution by ages - "+dataset_name+" dataset", "Emotion", save=True)
printDistribution(distributionByGender, "Emotions distribution by gender - "+dataset_name+" dataset", "Gender", save=True)
```

# Ages distribution #

```python
distribution=pd.Series(0, index=ageList)

for indexAge, age in enumerate(ageList):
    group = df.loc[df["age"] == indexAge]
    distribution[age] = group.shape[0]

print(distribution)
printPieChart(distribution, ageList, "Ages distribution - "+dataset_name+" dataset", save=True)
```
